With a threshold of 0.010, the accuracy is decreased a bit -> accuracy = 0.9652 
With a threshold of 0.020, the accuracy is decreased a bit -> accuracy = 0.9560
With a threshold of 0.015(mean of the two), the accuracy is -> 0.990740 (better than before)
Now testing for threshold of 0.016(might increase/decrease/constant) -> 0.990740
Now testing for threshold of 0.014(might increase/decrease/constant) -> 0.990740
Now testing for threshold of 0.017(might increase/decrease/constant) -> 0.990740
Now testing for threshold of 0.013                                   -> 0.96990740
As the value remains constant when 0.014<threshold<0.017, i will be taking the average value amongst it
So, threshold = 0.0155
